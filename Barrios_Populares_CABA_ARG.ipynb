```python
#Import libraries

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

*Data Exploration*

```python
#Charge and read data

current_path = os.getcwd()
data_folder_path = os.path.join(current_path, "data")
data_path = os.path.join(data_folder_path, "data.csv")
df = pd.read_csv(data_path)
df
```

```python
#See shape and columns
shape = df.shape
columns= df.columns

print("df shape:", shape)

print("list of",columns)
```

```python
#Index selection for our investigation
columns_to_explore = ['nombre_barrio', 'provincia','departamento', 'localidad',
                                'cantidad_viviendas_aproximadas',
                                'cantidad_familias_aproximada','anio_de_creacion',
                                'energia_electrica', 'efluentes_cloacales',
                                'agua_corriente', 'cocina', 'calefaccion', 'situacion_dominial',
                                'clasificacion_barrio', 'superficie_m2', 'personas_genero_masc',
                                'personas_genero_fem', 'personas_genero_otrx',
                                'recibe_asignacion_universal_por_hijo',
                                'tipo_propiedad_de_quienes_vieven_en_ella',
                                'tipo_propiedad_alquilada',
                                'tipo_propiedad_prestada']
```

```python
#Save the original data
data_frame_original = df
```

```python
#Create a new df for our investigation and show it
df= df[columns_to_explore]
df
```

```python
#Take the data of Ciudad Autónoma de Buenos Aires
lista_caba = ['Ciudad Autónoma de Buenos Aires']
caba = df[df['provincia'].isin(lista_caba)]
caba
```

*Data Cleaning*

```python
#We have a dataframe with 50 rows and 22 columns
#Now, we go to analize the data types
caba.dtypes
```

```python
#And see the null data
caba.isna().any().sum().sum()
caba = caba.drop('anio_de_creacion', axis=1)
```

```python
#We can see the column "anio de_creacion" have null data and it's not usefull for our investigation. 
#We proceed to clean it and check it
caba = caba.dropna()
caba.head(5)
```

```python
#We have this data in our dataframe
caba.info()
```

*Data Analysis*

What are the Barrios with the highest population density?
How many families live there?

```python
#We add the approximate number of homes
viviendas = caba.cantidad_viviendas_aproximadas.sum()
print("Cantidad aproximada de viviendas:", viviendas)
```

```python
#Now, we take the numbers of homes per "barrios" and see their disbrution
viviendas_por_barrio = caba.groupby(['nombre_barrio']).cantidad_viviendas_aproximadas.sum().sort_values(ascending=False)
print("Distibucion de hogares por", viviendas_por_barrio)
```

```python
#Add the number of families
cantidad_de_familias = caba.cantidad_familias_aproximada.sum()
print("Cantidad de familias aproximada:", cantidad_de_familias)
```

```python
#Distribution of families per barrio
caba.groupby(['nombre_barrio']).cantidad_familias_aproximada.sum().sort_values(ascending=False)
```

```python
#Barrio with the most families numbers
caba.sort_values('cantidad_familias_aproximada', ascending=False)[0:1]
```

In conclusion, we can see that the "barrio popular" with the most highest population density is "Padre Rodoflfo Ricardelli (Ex Villa 1-11-14). This barrio has approximately 14,000 homes and 15,4000 families live in it.

¿What types of Barrios Populares can we identify?

```python
caba.clasificacion_barrio.value_counts().plot.pie(autopct="%0.1f %%")
```

74% of the Barrios populaes are located in "Villas", while the remaining 26% are "Asentamiento".

*Asentamiento: Neighborhoods whose housing complexes seek to maintain the urban fabric as a continuity of the fabric of the formal city. The land is, for the most part, subdivided into plots that form blocks, respecting the layout of the streets. In many cases, places were reserved for equipment and green spaces. the density The population is usually smaller than in Villas.

*Villas: Neighborhoods whose housing complexes make up an irregular urban fabric, that is, they are not block neighborhoods. In these neighborhoods, traffic is mostly through narrow corridors and, especially in large cities, they tend to grow in height since there is little or no land availability. They have a high or medium population density and are generally located near production and consumption centers and on land near railroad tracks and water courses.

What is the relationship with the ownership of the property?

```python
caba.situacion_dominial.value_counts().plot.pie(autopct="%0.1f %%")
```

As can be seen, the majority of residents do not have a property title, purchase-sale ticket, or any other type of document that provides security of tenure.


How is the access to electric light services like?

```python
#access to the electrical grid
caba.energia_electrica.value_counts().plot.bar()
```

```python
#Electrical energy
caba.energia_electrica.value_counts().plot.pie(autopct="%0.1f %%")
```

Only 16% of electricity meters are in regular legal operation, compared to 72% that have an irregular connection to the electrical grid. The remaining percentage is measured among those who share the service as a community (10%) or because more than one family lives in the place (2%).

How are sewage effluents distributed?

```python
#Efluentes cloacales
caba.efluentes_cloacales.value_counts().plot.pie(autopct="%0.1f %%")
```

As we see, around 45% of the homes in these Barrios do not have access to the sewage effluent network with the health, pollution and mobility problems that this entails for the people who live there.

What is access to electricity and gas services like?

```python
#What is the access to the gas and electricity network?
plt.figure (figsize=(15,10))
sns.countplot(data = caba, x='cocina', hue='calefaccion')
plt.show
```

For cooking and heating, the majority of these popular neighborhoods depend on electricity, while a smaller percentage uses bottled gas and another uses firewood.

What is access to the water network like?

```python
#Agua correinte
caba.agua_corriente.value_counts().plot.pie(autopct="%0.1f %%")
```

92% of homes in Barrios Populares have an irregular connection to the water network. Only 2% have a legal connection and pay for that service
